```python
# 현재 작업 여기서
conn = sqlite3.connect('data/dbfiles/TOTAL.db')
cur = conn.cursor()
cur.execute("ATTACH DATABASE 'data/dbfiles/RATINGS.db' AS t2")
cur.execute(
    """
    SELECT t1.id, t1.release_date, t2.rating
    FROM TOTAL t1
    JOIN RATINGS ON t1.id = t2.id
    WHERE t1.release_date BETWEEN '1990-01-01' AND '1999-12-31'
    ORDER BY t2.rating DESC;
    """
)
results = cur.fetchall()
print(results)
```
